```python
mode=round(titanic_df['age'].mode()[0])
```

```python
titanic_df['age']=titanic_df['age'].fillna(mode)
```

```python
titanic_df['deck']=titanic_df['deck'].fillna('C')
```

Observations

titanic_df.info() shows 891 rows and 15 columns.

Some columns like age, deck, and embarked contain missing values.

describe() shows:

Average passenger age ≈ 30 years.

Mean fare ≈ 32 (with high variation).

Survival rate: ~38% (from .value_counts() on survived).

More males (577) than females (314).

Most passengers were in 3rd class.

```python


# Pairplot for numerical columns
sns.pairplot(titanic_df, vars=['age', 'fare'], hue='survived')
plt.show()


```

Pairplot:

Survivors generally paid higher fares and were slightly younger.

Passengers with low fares and older age had lower survival rates.

```python
# Correlation heatmap
sns.heatmap(titanic_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

Heatmap:

fare and class are negatively correlated (higher class → higher fare).

survived correlates positively with fare and pclass (1st class more likely to survive).

```python
sns.histplot(titanic_df['age'], bins=30, kde=True)
plt.title("Distribution of Age")
plt.show()
```

Histogram: Most passengers were between 20–40 years old. Few children or elderly

```python
# Boxplot of Fare by Class
sns.boxplot(x='class', y='fare', data=titanic_df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

Boxplot:

Median fare increases from 3rd → 1st class.

1st class has many high-value outliers (expensive tickets).

```python

# Scatterplot of Age vs Fare by Survival
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic_df)
plt.title("Age vs Fare by Survival")
plt.show()
```

Scatterplot:

Survivors tend to cluster around higher fares.

Non-survivors are more common in lower fare regions.
